```python
import pandas as pd
```

```python
df = pd.read_csv('AiO.tsv', sep='\t')
```

```python
df.head()
```

```python
#results
#0: lr
#1: epochs
#2: train evaluation (loss, acc)
#3: validation evalutaion (loss, acc)
#4: test evaluation (loss, acc)
```

```python
example = "0.0001 [16, [0.3887983560562134, 0.8352941274642944], [0.48602333664894104, 0.800000011920929], [0.7043976783752441, 0.6700000166893005]]"
```

```python
def splitter(s):
    arr = s.split('[', 1) #arr[0] #space might be missing
    x = eval('[' + arr[1]) #x[0], x[1][1], x[2][1], x[3][1]
    return {'lr': arr[0], 'epochs': x[0], 'train_acc': x[1][1], 'val_acc': x[2][1], 'test_acc': x[3][1]}
```

```python
splitter(example)
```

```python
df_ = df.apply(lambda x: splitter(x.Results), axis='columns', result_type='expand')
```

```python
df_.head()
```

```python
df = pd.concat([df, df_], axis='columns')
```

```python
df.head()
```

```python
df.groupby(['Modell', 'LearningType', 'LearningRate'])['test_acc'].max()
```

```python
df_res = df.groupby(['Modell', 'LearningType', 'LearningRate'])['test_acc'].max().reset_index()
```

```python
df_tmp = df.groupby(['Modell', 'LearningType', 'LearningRate'])['test_acc'].mean().reset_index()
df_tmp['test_acc_avg'] = df_tmp['test_acc']
df_tmp.drop(['Modell', 'LearningType', 'LearningRate', 'test_acc'], axis='columns', inplace=True)
df_res = pd.concat([df_res, df_tmp], axis='columns')
```

```python
df_tmp = df.groupby(['Modell', 'LearningType', 'LearningRate'])['test_acc'].std().reset_index()
df_tmp['test_acc_std'] = df_tmp['test_acc']
df_tmp.drop(['Modell', 'LearningType', 'LearningRate', 'test_acc'], axis='columns', inplace=True)
df_res = pd.concat([df_res, df_tmp], axis='columns')
```
